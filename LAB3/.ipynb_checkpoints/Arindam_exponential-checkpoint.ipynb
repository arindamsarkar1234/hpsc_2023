```python
import numpy as np
import math
import matplotlib.pyplot as plt
import timeit


x=np.linspace(0,100,1000)
z=np.exp(x)




def arindamexp(x, n=100):
    sum = 1.0
    for i in range (1, n+1):
         sum=sum+(x**i/math.factorial(i))
    return sum

y = arindamexp(x)

plt.plot(x,z)
plt.plot(x,y)
plt.show()



%timeit arindamexp(4)

```
